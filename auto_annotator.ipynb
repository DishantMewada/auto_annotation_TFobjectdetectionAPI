```python
# Python program to create
# a file explorer in Tkinter
  
# import all components
# from the tkinter library
from tkinter import *
  
# import filedialog module
from tkinter import filedialog
  
# Function for opening the
# file explorer window
def browseFiles():
    filename = filedialog.askopenfilename(initialdir = "/",
                                          title = "Select a File",
                                          filetypes = (("Text files",
                                                        "*.jpg*"),
                                                       ("all files",
                                                        "*.*")))
      
    # Change label contents
    label_file_explorer.configure(text="File Opened: "+filename)
      
      
                                                                                                  
# Create the root window
window = Tk()
  
# Set window title
window.title('File Explorer')
  
# Set window size
window.geometry("500x500")
  
#Set window background color
window.config(background = "white")
  
# Create a File Explorer label
label_file_explorer = Label(window,
                            text = "File Explorer using Tkinter",
                            width = 100, height = 4,
                            fg = "blue")
  
      
button_explore = Button(window,
                        text = "Browse Files",
                        command = browseFiles)
  
button_exit = Button(window,
                     text = "Exit",
                     command = exit)
  
# Grid method is chosen for placing
# the widgets at respective positions
# in a table like structure by
# specifying rows and columns
label_file_explorer.grid(column = 1, row = 1)
  
button_explore.grid(column = 1, row = 2)
  
button_exit.grid(column = 1,row = 3)
  
# Let the window wait for any events
window.mainloop()
```

```python
import tkinter as tk

root = tk.Tk()

v = tk.IntVar()
v.set(1)  # initializing the choice, i.e. Python

languages = [("Python", 101),
   	     ("Perl", 102),
    	     ("Java", 103),
             ("C++", 104),
             ("C", 105)]

def ShowChoice():
    print(v.get())

tk.Label(root, 
         text="""Choose your favourite 
programming language:""",
         justify = tk.LEFT,
         padx = 20).pack()

for language, val in languages:
     tk.Radiobutton(root, 
                  text=language,
                  indicatoron = 0,
                  width = 20,
                  padx = 20, 
                  variable=v, 
                  command=ShowChoice,
                  value=val).pack(anchor=tk.W)



root.mainloop()
```

```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog
import tkinter as tk

import os

gui = Tk()
gui.geometry("1000x500")
gui.title("Auto-annotation")

finalEntry = {}

prerequisite = Label(gui, text = '0. Install Prerequisite',pady = 10)
prerequisite.grid(row=0,column=0)


def getModel():
    os.system("git clone https://github.com/tensorflow/models.git && cd models && git checkout -f e04dafd04d69053d3733bb91d47d0d95bc2c8199")
    downloadModelStatus = Label(gui, text = 'Complete')
    downloadModelStatus.grid(row=1,column=1)
    
downloadModel = ttk.Button(gui, text="Download Tf-Model",command=getModel,width=20)
downloadModel.grid(row=0,column=1)

downloadModelStatus = Label(gui, text = '')
downloadModelStatus.grid(row=1,column=1)

####

def compileModelFunc():
    os.system("cd ./models/research && protoc object_detection/protos/*.proto --python_out=. && python setup.py build && python setup.py install")
    compileModelStatus = Label(gui, text = 'Complete')
    compileModelStatus.grid(row=1,column=2)
    
compileModel = ttk.Button(gui, text="Compile Model",command=compileModelFunc,width=20)
compileModel.grid(row=0,column=2)

compileModelStatus = Label(gui, text = '')
compileModelStatus.grid(row=1,column=2)

####

path = os.getcwd()

workspace_path = path+"/workspace"
pretrained_model_path = workspace_path+"/pre-trained-models"

def buildDirFunc():
    try:
        os.mkdir(workspace_path)
        os.mkdir(pretrained_model_path)
    except OSError:
        buildDirStatus = Label(gui, text = 'OS Error')
        buildDirStatus.grid(row=1,column=3)
    else:
        buildDirStatus = Label(gui, text = 'Complete')
        buildDirStatus.grid(row=1,column=3)
    
buildDir = ttk.Button(gui, text="Build Dirs",command=buildDirFunc,width=20)
buildDir.grid(row=0,column=3)

buildDirStatus = Label(gui, text = '')
buildDirStatus.grid(row=1,column=3)

####

def installPIPfunc():
    os.system("pip3 install wheel tensorflow-gpu==2.5.0 tf-models-official matplotlib Cython")
    buildPIPStatus = Label(gui, text = 'Complete')
    buildPIPStatus.grid(row=1,column=4)
    
buildPIP = ttk.Button(gui, text="Install PIPs",command=installPIPfunc,width=20)
buildPIP.grid(row=0,column=4)

buildPIPStatus = Label(gui, text = '')
buildPIPStatus.grid(row=1,column=4)

##############################################################
displayPath = Label(gui, text = '/Folder/Path/', width=50,pady = 15)
displayPath.grid(row=2,column=1,columnspan=5)

radioValue = IntVar() 

def getFolderPath():
    folder_selected = filedialog.askdirectory()
    folderPath.set(folder_selected)
    folder = folderPath.get()
    displayPath.configure(text = folder)
    finalEntry["imagesFolder"] = folder


folderPath = StringVar()
importPath = ttk.Button(gui, text="1. Import Folder",command=getFolderPath)
importPath.grid(row=2,column=0)


##############################################################

zoo = [("EfficientDet D0 512x512", "https://download.tensorflow.org/models/object_detection/tf2/20200711/efficientdet_d0_coco17_tpu-32.tar.gz",4),
       ("EfficientDet D1 640x640", "https://download.tensorflow.org/models/object_detection/tf2/20200711/efficientdet_d1_coco17_tpu-32.tar.gz",5),
       ("EfficientDet D2 768x768", "https://download.tensorflow.org/models/object_detection/tf2/20200711/efficientdet_d2_coco17_tpu-32.tar.gz",6),
       ("EfficientDet D3 896x896", "https://download.tensorflow.org/models/object_detection/tf2/20200711/efficientdet_d3_coco17_tpu-32.tar.gz",7),
       ("EfficientDet D4 1024x1024", "https://download.tensorflow.org/models/object_detection/tf2/20200711/efficientdet_d4_coco17_tpu-32.tar.gz",8),
       ("SSD MobileNet V1 FPN 640x640", "https://download.tensorflow.org/models/object_detection/tf2/20200711/ssd_mobilenet_v1_fpn_640x640_coco17_tpu-8.tar.gz",9),
       ("SSD MobileNet V2 FPNLite 320x320", "https://download.tensorflow.org/models/object_detection/tf2/20200711/ssd_mobilenet_v2_fpnlite_320x320_coco17_tpu-8.tar.gz",10)
      ]

v = tk.IntVar()

def ShowChoice():
    #print(v.get())
    for model, link, index in zoo:
        if index == v.get():
            finalEntry["downloadLink"] = link

tk.Label(gui, 
         text="""2. Choose object detection pretrained-model""",
         justify = tk.LEFT,
         pady = 10).grid(row=3,column=0)


for model, link, index in zoo:
    tk.Radiobutton(gui, 
                   text=model,
                   #indicatoron = 0,
                   width = 30,
                   padx = 20, 
                   variable=v, 
                   command=ShowChoice,
                   value=index).grid(row=index,column=0)
#####################################################################
displayDownloadStatus = Label(gui, text = '', width=50,pady = 15)
displayDownloadStatus.grid(row=11,column=1,columnspan=5)


def downloadPretrainedModelFunc():
    
    tarName = finalEntry.get('downloadLink').split("/")
    folderTAR = tarName[-1].split(".")
    print(finalEntry)
    tarNamePath = './workspace/pre-trained-models/{0}'.format(tarName[-1])

    if os.path.isfile(tarNamePath):
        pass
    else:
        stringForCMD = "cd ./workspace/pre-trained-models && wget --no-check-certificate {0} && tar -zxvf {1}".format(finalEntry.get('downloadLink'),tarName[-1])
        os.system(stringForCMD)
    
    displayDownloadStatus = Label(gui, text = 'Complete', width=50,pady = 15)
    displayDownloadStatus.grid(row=11,column=1,columnspan=5)



downloadPretrainedModel = ttk.Button(gui, text="3. Download the model",command=downloadPretrainedModelFunc)
downloadPretrainedModel.grid(row=11,column=0)


gui.mainloop()
```
